```python
df = pd.read_csv("../input/hr-employee-attrition/HR Employee Attrition.csv")
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
new_df=df.drop('Attrition',axis=1)
```

```python
new_df.head()
```

```python
s = (new_df.dtypes=='object')
object_cols = list(s[s].index)

print("Categorical: ")
print(object_cols)
```

```python
new_df[['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'Over18', 'OverTime']].nunique()
```

```python
y = df['Attrition']
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder

X=new_df
X_train, X_valid, y_train, y_valid = train_test_split(X,y)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
enc = OneHotEncoder(handle_unknown='ignore',sparse=False)
OH_cols_train = pd.DataFrame(enc.fit_transform(X_train[object_cols]))
OH_cols_valid = pd.DataFrame(enc.transform(X_valid[object_cols]))
OH_cols_train.index = X_train.index
OH_cols_valid.index = X_valid.index
num_X_train = X_train.drop(object_cols,axis=1)
num_X_valid = X_valid.drop(object_cols,axis=1)
OH_X_train = pd.concat([num_X_train, OH_cols_train],axis=1)
OH_X_valid = pd.concat([num_X_valid, OH_cols_valid],axis=1)
```

```python
OH_X_train.head()
```

```python
OH_X_train.shape
```

```python
y_valid.values.reshape(1,-1).shape
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()


```

```python
new_y_train = le.fit_transform(y_train)
new_y_train.shape
new_y_valid = le.transform(y_valid)
new_y_valid.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor(max_depth=5)
reg.fit(OH_X_train,new_y_train)
```

```python
results = reg.predict(OH_X_valid)
```

```python
results.shape
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(new_y_valid, results)
```
